```python
import pandas as pd
```

```python
apps = pd.read_csv("datasets/apps.csv")
for char in [",", "+"]:
    apps["Installs"] = apps["Installs"].apply(lambda s: s.replace(char, ""))
apps["Installs"] = apps["Installs"].astype("int")
apps.info()
```

```python
app_category_info = apps.groupby("Category").agg(
    {"App": "count", "Price": "mean", "Rating": "mean"})
app_category_info.rename(columns={
                         "App": "Number of apps",
                         "Price": "Average price",
                         "Rating": "Average rating"}
                         )
app_category_info.head()
```

```python
free_finance_apps = apps[(apps["Category"] == "FINANCE")
                         & (apps["Type"] == "Free")]
free_finance_apps.head()
```

```python
user_reviews = pd.read_csv("datasets/user_reviews.csv")
average_user_review_by_app = user_reviews.groupby(
    "App").agg({"Sentiment Score": "mean"})
average_user_review_by_app.head()
```

```python
top_free_finance_apps_and_user_reviews = free_finance_apps.merge(
    average_user_review_by_app, on="App").nlargest(10, "Sentiment Score")
top_10_user_feedback = average_user_review_by_app[average_user_review_by_app.index.isin(
    top_free_finance_apps_and_user_reviews["App"])].nlargest(10, "Sentiment Score")
top_10_user_feedback
```
